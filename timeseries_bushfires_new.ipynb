```python
df1 =  df[["acq_date","Sub Urban_x","State","frp",'maxtempC', 'mintempC', 'totalSnow_cm',
       'sunHour', 'uvIndex', 'uvIndex.1', 'moon_illumination', 'moonrise',
       'moonset', 'sunrise', 'sunset', 'DewPointC', 'FeelsLikeC', 'HeatIndexC',
       'WindChillC', 'WindGustKmph', 'cloudcover', 'humidity', 'precipMM',
       'pressure', 'tempC', 'visibility', 'winddirDegree', 'windspeedKmph']]
```

```python
grouped_multiple = df1.groupby(['date', 'Sub Urban_x',"State"]).mean().reset_index()
print(grouped_multiple)
```
