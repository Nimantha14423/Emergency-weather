```
import pandas as pd
from glob import glob
```

```
from google.colab import files
uploaded = files.upload()
```

```
li =[]
```

```
for filename in uploaded:
  df = pd.read_csv(filename, skiprows=[0], index_col=None,header=None)
  li.append(df)
```

```
weather = pd.concat(li, axis=0, ignore_index=True)
weather.columns =['date','maxtempC','mintempC','totalsnow','sunHour','uvindex','moon_illumination','moonrise','moonset','sunrise','sunset','DewPointC','FeelsLikeC','HeatIndexC','WindChillC','WindGustKmph','cloudcover','humidity','precipMM','pressure','tempC','visibility','winddirDegreeI','windspeedKmph','location']
```
