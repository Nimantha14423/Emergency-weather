```python
import pandas as pd
from glob import glob
```

```python
fire_data = sorted(glob('F:/Masters of Data Science/Trimester 3/team project B/bushfire/viirs_201*.csv'))
```

```python
df_we = pd.concat((pd.read_csv(file).assign(filename = file)
          for file in fire_data), ignore_index = True)
```

```python
df_we
```

```python
df_we["latitude"] = df_we.latitude.astype(str)
df_we["longitude"] = df_we.longitude.astype(str)
```

```python
longlat = df_we[['latitude','longitude']].apply(tuple, axis=1)
longlat
```

```python
import reverse_geocoder as revgeo
```

```python
longlat = longlat.tolist()
```

```python
results = revgeo.search(longlat)
```

```python
lonldf = pd.DataFrame(results)
```

```python
df_we['city'] = lonldf['name']
```
